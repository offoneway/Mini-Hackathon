```python
# ใช้ URL มาเพื่อโหลดไฟล์
file_url = 'https://storagepracticebdidev.blob.core.windows.net/dev-uploads/app/bdi/uploads1-9-20260312-162658503.csv?sv=2026-02-06&st=2026-03-12T09%3A26%3A58Z&se=2028-12-06T09%3A26%3A58Z&sr=b&sp=r&sig=fV6wspFvE9dTgkzY9xLbZ%2FXP4i%2B%2FN0BliGZAx9%2Bzywo%3D'
df_elec = pd.read_csv(file_url, encoding='utf-8-sig')
print('✅ โหลดไฟล์ electric.csv จาก URL สำเร็จแล้ว')


print('=' * 60)
print('📅 ข้อมูลสถิติหน่วยจำหน่ายไฟฟ้าแยกตามเขต')
print(f'   ➜ จำนวนแถว: {df_elec.shape[0]:,} แถว | คอลัมน์: {df_elec.shape[1]} คอลัมน์')
print(f'   ➜ คอลัมน์: {df_elec.columns.tolist()}')
print('=' * 60)

print()
print('🔍 ตัวอย่างสถิติหน่วยจำหน่ายไฟฟ้าแยกตามเขต:')
display(df_elec)
```

```python
#เติมเลขปีใน column ปี เเละเปลี่ยน datatype เป็น int
df_elec['ปี'] = df_elec['ปี'].ffill()
df_elec['ปี'] = df_elec['ปี'].astype(int)

print('✅ เติมข้อมูลในคอลัมน์ \'ปี\' สำเร็จแล้ว')
print('\n🔍 ตัวอย่างข้อมูลหลังเติมปี:')
display(df_elec[['ปี','(หักไฟทำการ)']])
```

```python
#เปลี่ยน datatype ค่าหักไฟทำการเป็น float
df_elec['(หักไฟทำการ)'] = df_elec['(หักไฟทำการ)'].str.replace(',', '', regex=False).astype(float)
df_elec['(หักไฟทำการ)'].dtype
```

```python
# 1. ทำความสะอาดชื่อคอลัมน์ โดยลบช่องว่างนำหน้า/ต่อท้าย
df_elec.columns = df_elec.columns.str.strip()

# 2. สร้างคอลัมน์ 'Month_num' จากชื่อเดือนภาษาไทยเเละลบช่องที่เว้นว่างในเดือน
month_mapping = {
    'ม.ค.': 1, 'ก.พ.': 2, 'มี.ค.': 3, 'เม.ย.': 4, 'พ.ค.': 5, 'มิ.ย.': 6,
    'ก.ค.': 7, 'ส.ค.': 8, 'ก.ย.': 9, 'ต.ค.': 10, 'พ.ย.': 11, 'ธ.ค.': 12
}
df_elec['Month_num'] = df_elec['เดือน'].str.strip().map(month_mapping)
display(df_elec[['ปี','Month_num']])
```

```python

# 3. สร้างคอลัมน์ 'Date' ในรูปแบบ datetime โดยแปลงปีพุทธศักราชเป็นคริสต์ศักราชชั่วคราวเพื่อสร้างวันที่
df_elec['Date'] = pd.to_datetime((df_elec['ปี'].astype(int) - 543).astype(str) + '-' + df_elec['Month_num'].astype(int).astype(str) + '-01')

print('✅ การเตรียมข้อมูลสำเร็จแล้ว')
print('\n🔍 ตัวอย่างข้อมูลหลังจากเตรียม:')
display(df_elec[['Date', '(หักไฟทำการ)']])
```

### 🔎2. เขตไหนใช้ไฟมากที่สุด และใครโตเร็วที่สุด?

จากกราฟด้านล่าง สังเกตได้ถึงสิ่งเหล่านี้:

1.  **ปริมาณการใช้ไฟฟ้า**: เเต่ละเขตมีการใช้ปริมาณมากน้อยเเค่ไหน ในช่วงเวลาไหนบ้าง?
2.  **อัตราการเติบโตของการใช้ไฟฟ้า**: เเต่ละเขตมีอัตราการเติบโตของการใช้ไฟฟ้าที่มากขึ้น หรือน้อยลงโดยเปรียบเทียบปีแรก (2562) และปีสุดท้าย (2567) ที่ข้อมูลครบล่าสุด จากการใช้ไฟฟ้าทั้งหมดของเเต่ละเขต



```python
# 1. กำหนดคอลัมน์ของเขตการจำหน่ายไฟฟ้าทั้ง 18 เขต
district_cols = ['วัดเลียบ', 'คลองเตย', 'ยานนาวา', 'บางกะปิ', 'มีนบุรี', 'สมุทรปราการ', 'บางพลี', 'สามเสน', 'นนทบุรี', 'บางใหญ่', 'ธนบุรี', 'ราษฎร์บูรณะ', 'บางขุนเทียน', 'บางเขน', 'บางนา', 'บางบัวทอง', 'ลาดกระบัง', 'นวลจันทร์']

# 2. คำนวณปริมาณการใช้ไฟฟ้าทั้งหมด (Total Consumption) สำหรับแต่ละเขต
# โดยการรวมค่าทั้งหมดในแต่ละคอลัมน์ของเขต
total_consumption_by_district = df_elec[district_cols].sum().reset_index()
total_consumption_by_district.columns = ['District', 'Total_Consumption']

# 3. คำนวณอัตราการเติบโต (Growth Rate) สำหรับแต่ละเขต
# ใช้ข้อมูลจากปีแรก (2562) และปีสุดท้าย (2567) ที่เป็นปีเต็ม

# ตรวจสอบปีที่มีข้อมูลก่อน
years_in_data = df_elec['ปี'].unique()
start_year_buddhist = years_in_data.min()
end_year_buddhist = years_in_data.max()

# ถ้าปีสุดท้ายมีข้อมูลไม่ครบเดือน อาจใช้ปีก่อนหน้า หรือใช้ max เพื่อให้ได้ข้อมูลล่าสุด
# ในกรณีนี้เราจะใช้ปีแรก 2562 (2019) และปีสุดท้าย 2567 (2024) เนื่องจากปี 2568 อาจไม่เต็มปี
consumption_2562 = df_elec[df_elec['ปี'] == start_year_buddhist][district_cols].sum()
consumption_2567 = df_elec[df_elec['ปี'] == 2567][district_cols].sum()

growth_rate_by_district = ((consumption_2567 - consumption_2562) / consumption_2562) * 100
growth_rate_by_district = growth_rate_by_district.reset_index()
growth_rate_by_district.columns = ['District', 'Growth_Rate']

# 4. รวมข้อมูล Total Consumption และ Growth Rate เข้าด้วยกัน
df_district_analysis = pd.merge(total_consumption_by_district, growth_rate_by_district, on='District')

print('✅ ข้อมูลสำหรับการวิเคราะห์เขตเสร็จสมบูรณ์แล้ว')
print('\n🔍 ตัวอย่างข้อมูลวิเคราะห์แต่ละเขต:')
display(df_district_analysis.head())

# 5. สร้างกราฟ Scatter Plot
fig = px.scatter(
    df_district_analysis,
    x='Total_Consumption',
    y='Growth_Rate',
    text='District',
    size='Total_Consumption', # ขนาดของจุดแสดงตามปริมาณการใช้ไฟฟ้า
    color='District', # แยกสีตามเขต
    title='ปริมาณการใช้ไฟฟ้าและอัตราการเติบโตของแต่ละเขต (พ.ศ. 2562 - 2567)',
    labels={
        'Total_Consumption': 'ปริมาณการใช้ไฟฟ้าทั้งหมด (GWh)',
        'Growth_Rate': '''อัตราการเติบโต (%)
        (พ.ศ. 2562 เทียบ พ.ศ. 2567)'''
    },
    height=600
)

fig.update_traces(textposition='top center')
fig.update_layout(hovermode="x unified")
fig.show()
```

จากกราฟ Scatter Plot 'ปริมาณการใช้ไฟฟ้าและอัตราการเติบโตของแต่ละเขต (พ.ศ. 2562 - 2567)' เราสามารถระบุเขตต่างๆ ได้ดังนี้ครับ:

**1. เขตที่ใช้ไฟฟ้ามากและยังมีการเติบโตต่อเนื่อง (High Consumption & Growing):**

*   **บางพลี:** เป็นเขตที่มีปริมาณการใช้ไฟฟ้าทั้งหมดสูงที่สุดในบรรดาเขตทั้งหมด (อยู่ที่ประมาณ 36,452 GWh) และยังคงมีอัตราการเติบโตที่เป็นบวก (ประมาณ 6.68%) แสดงให้เห็นว่าเป็นเขตที่มีกิจกรรมทางเศรษฐกิจสูงและยังคงขยายตัว
*   **คลองเตย, นนทบุรี, สามเสน, บางเขน, บางนา:** เขตเหล่านี้มีปริมาณการใช้ไฟฟ้าที่สูงเช่นกัน และมีอัตราการเติบโตที่เป็นบวก แสดงให้เห็นถึงการขยายตัวทางเศรษฐกิจอย่างต่อเนื่อง

**2. เขตที่ใช้ไฟฟ้าน้อยแต่มีอัตราการเติบโตที่รวดเร็ว (Low Consumption & Fast Growing):**

*   **บางบัวทอง:** เป็นเขตที่มีปริมาณการใช้ไฟฟ้าไม่สูงมากนัก (ประมาณ 13,321 GWh) แต่มีอัตราการเติบโตที่โดดเด่นและสูงที่สุด (ประมาณ 28.91%) บ่งชี้ว่าเขตนี้มีการพัฒนาอย่างรวดเร็วและมีศักยภาพในการขยายตัวสูงในอนาคต
*   **บางใหญ่, มีนบุรี, ลาดกระบัง, ธนบุรี:** เขตเหล่านี้มีปริมาณการใช้ไฟฟ้าอยู่ในระดับปานกลางถึงน้อย แต่มีอัตราการเติบโตที่ค่อนข้างสูง (ตั้งแต่ประมาณ 9% ถึง 17%) ซึ่งแสดงถึงการพัฒนาและขยายตัวของพื้นที่เหล่านั้นครับ

**ข้อสังเกตเพิ่มเติม:** บางเขตอย่างเช่น สมุทรปราการ และ ราษฎร์บูรณะ แม้จะมีปริมาณการใช้ไฟฟ้าสูง แต่กลับมีอัตราการเติบโตที่เป็นลบ ซึ่งอาจบ่งชี้ถึงความท้าทายหรือการชะลอตัวทางเศรษฐกิจในพื้นที่นั้นๆ ครับ

```python
# 1. เตรียมข้อมูลสำหรับวิเคราะห์รูปแบบฤดูกาล
# เลือกเฉพาะคอลัมน์ที่เกี่ยวข้องกับการใช้ไฟฟ้าของแต่ละเขตและคอลัมน์ 'Month_num'
df_seasonal = df_elec[['Month_num'] + district_cols]

# Group by 'Month_num' และคำนวณค่าเฉลี่ยการใช้ไฟฟ้าสำหรับแต่ละเขตในแต่ละเดือน
df_seasonal_avg = df_seasonal.groupby('Month_num')[district_cols].mean().reset_index()

# แปลงข้อมูลจาก wide format เป็น long format เพื่อใช้กับ Plotly Express ได้ง่ายขึ้น
df_seasonal_avg_melted = df_seasonal_avg.melt(id_vars='Month_num', var_name='District', value_name='Average_Consumption')

# สร้างชื่อเดือนภาษาไทยเพื่อแสดงผลบนกราฟ
month_names = {
    1: 'ม.ค.', 2: 'ก.พ.', 3: 'มี.ค.', 4: 'เม.ย.', 5: 'พ.ค.', 6: 'มิ.ย.',
    7: 'ก.ค.', 8: 'ส.ค.', 9: 'ก.ย.', 10: 'ต.ค.', 11: 'พ.ย.', 12: 'ธ.ค.'
}
df_seasonal_avg_melted['Month'] = df_seasonal_avg_melted['Month_num'].map(month_names)

print('✅ ข้อมูลรูปแบบฤดูกาลเฉลี่ยต่อเดือนของแต่ละเขตสำเร็จแล้ว')
print('\n🔍 ตัวอย่างข้อมูล:')
display(df_seasonal_avg_melted.head())

# 2. สร้างกราฟเส้นเพื่อเปรียบเทียบรูปแบบฤดูกาล
fig = px.line(
    df_seasonal_avg_melted,
    x='Month',
    y='Average_Consumption',
    color='District',
    line_group='District',
    title='รูปแบบฤดูกาลของการใช้ไฟฟ้าเฉลี่ยรายเดือนของแต่ละเขต',
    labels={
        'Month': 'เดือน',
        'Average_Consumption': 'ปริมาณการใช้ไฟฟ้าเฉลี่ย (GWh)',
        'District': 'เขต'
    },
    height=600,
    hover_name='District'
)

fig.update_xaxes(categoryorder='array', categoryarray=list(month_names.values()))
fig.update_layout(hovermode="closest")
fig.show()
```

จากกราฟเส้น 'รูปแบบฤดูกาลของการใช้ไฟฟ้าเฉลี่ยรายเดือนของแต่ละเขต' ที่สร้างไว้ เราสามารถระบุลักษณะการใช้ไฟฟ้าตามฤดูกาลของแต่ละเขต พร้อมตั้งสมมติฐานได้ดังนี้ครับ:

**1. เขตที่มีความผันผวนตามฤดูกาลสูงผิดปกติ:**

*   **เขตที่สังเกตได้ชัด:** เขตที่มีกราฟพุ่งสูงขึ้นอย่างเห็นได้ชัดในช่วงเดือนมีนาคม-พฤษภาคม และลดลงในช่วงเดือนอื่น ๆ ได้แก่ **บางพลี, สมุทรปราการ, ลาดกระบัง, บางขุนเทียน, บางนา, มีนบุรี, บางใหญ่, บางบัวทอง** (โดยเฉพาะบางพลีและสมุทรปราการจะมีปริมาณการใช้สูงที่สุดและมีรูปแบบคล้ายกัน)
*   **ลักษณะการใช้พลังงานแบบใด:**
    *   **สมมติฐาน: เขตอุตสาหกรรม/พาณิชยกรรม และที่อยู่อาศัยที่หนาแน่น** การที่การใช้ไฟฟ้าพุ่งสูงขึ้นในช่วงหน้าร้อน (มี.ค.-พ.ค.) อย่างมีนัยสำคัญ บ่งชี้ถึงการใช้เครื่องปรับอากาศจำนวนมากในภาคครัวเรือนและภาคธุรกิจ นอกจากนี้ในเขตอุตสาหกรรมที่ต้องมีการควบคุมอุณหภูมิในกระบวนการผลิต อาจมีการใช้ไฟฟ้าเพิ่มขึ้นในฤดูร้อนเช่นกัน เขตเหล่านี้อาจมีโรงงาน สถานประกอบการขนาดใหญ่ หรือแหล่งที่อยู่อาศัยที่มีการใช้เครื่องปรับอากาศจำนวนมาก

**2. เขตที่มีการใช้ไฟฟ้าคงที่ตลอดปี:**

*   **เขตที่สังเกตได้ชัด:** เขตที่กราฟค่อนข้างแบนราบ ไม่มีความผันผวนขึ้นลงตามฤดูกาลมากนัก แต่ยังคงมีปริมาณการใช้ไฟฟ้าค่อนข้างสูง ได้แก่ **คลองเตย, ยานนาวา, สามเสน, วัดเลียบ, ธนบุรี, ราษฎร์บูรณะ, นนทบุรี, บางกะปิ**
*   **ลักษณะการใช้พลังงานแบบใด:**
    *   **สมมติฐาน: เขตพาณิชยกรรม/ธุรกิจ/ราชการ/ที่อยู่อาศัยแบบตึกสูง** เขตเหล่านี้มักเป็นศูนย์กลางธุรกิจ, ออฟฟิศ, ห้างสรรพสินค้า, โรงพยาบาล หรืออาคารที่พักอาศัยขนาดใหญ่ที่มีการใช้ไฟฟ้าต่อเนื่องตลอดทั้งปี ไม่ว่าจะฤดูไหนก็ตาม การใช้เครื่องปรับอากาศในอาคารเหล่านี้อาจถูกควบคุมในระดับหนึ่ง หรือมีระบบระบายความร้อนที่ประสิทธิภาพสูง ทำให้ความต้องการไฟฟ้าไม่ผันผวนตามอุณหภูมิภายนอกมากนักเมื่อเทียบกับเขตที่เน้นที่อยู่อาศัยทั่วไป หรืออุตสาหกรรมที่อ่อนไหวต่ออุณหภูมิ
